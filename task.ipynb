```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tip(in).csv')
print(df.head())
```

```python
print(df.shape)
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
# line chart
import matplotlib.pyplot as plt
plt.plot(df['time'],df['total_bill'])
plt.xlabel('time')
plt.ylabel('total_bill')
plt.title('line chart')
plt.show()
plt.show()
```

```python
# bar graph
plt.bar(df['sex'],df['total_bill'],color='red')
plt.xlabel('time')
plt.ylabel('total_bill')
plt.title('category competition')
plt.show()
```

```python
# histogram
plt.hist(df['total_bill'],bins=10,color='green',edgecolor='blue')
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.title('value distribution')
plt.show()
```

```python
# scatterplot
plt.scatter(df['total_bill'],df['tip'],color='purple')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.title('scatterplot')
plt.show()
```

```python
# outliers and spread
plt.boxplot(df['total_bill'])
plt.ylabel('total_bill')
plt.title('boxplot')
plt.show()
```

```python
# pie chart
labels=['smoker','non-smoker']
sizes=[df['smoker'].value_counts()[0],df['smoker'].value_counts()[1]]
colors=['red','blue']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%')
plt.show()
```

```python
# subnplots
plt.subplot(2,2,1)
plt.bar(df['sex'],df['total_bill'],color='red')
plt.xlabel('time')
plt.ylabel('total_bill')
plt.title('category competition')

plt.subplot(2,2,2)
plt.hist(df['total_bill'],bins=10,color='green',edgecolor='blue')
plt.xlabel('total_bill')
plt.ylabel('frequency')
plt.title('value distribution')
```
