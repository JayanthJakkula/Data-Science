Using the existing DataFrame, perform the following operations:

âœ… Print the details (name, age, marks) of the student who scored the highest marks.

```python
s1=pd.Series(['jayanth','viswas','karan'])
s2=pd.Series([23,27,25])
s3=pd.Series([90,50,95])
d={'name':s1,'age':s2,'marks':s3}
s=pd.DataFrame(d)
s
print()
print("Highest marks student details:")
print(s.loc[s['marks'].idxmax()])

print()
print("Add new student")
s.loc[3]=['chandrakanth',24,70]
print(s)

print()
print("Display the details of students who are 25 years old or older.")
print(s.loc[s['age']>25])

print()
print("Increase the marks of students below 80 by 10 bonus points.")
s.loc[s['marks']<80,'marks']+=10
print(s)
print()
print(" Calculate and print the average marks of all students after the bonus.")
print(s['marks'].mean())
```

```python
df=pd.read_csv('/content/drive/My Drive/OnlineRetail.csv',encoding='unicode_escape')
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows where CustomerID is missing (canâ€™t link to a customer)
df = df.dropna(subset=['CustomerID'])
```

```python
# Fill missing descriptions or other columns
df['Description'] = df['Description'].fillna("Unknown")
print(df)
```

```python
# Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df
df.isnull().sum()

```

```python
# Drop records where date parsing failed
df = df.dropna(subset=['InvoiceDate'])
df
```

```python
returns = df[df['Quantity'] < 0]
print("Returns count:", len(returns))
```

```python
df = df[df['Quantity'] > 0]
df = df[df['UnitPrice'] > 0]
```

```python

print("Duplicates:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
```

```python
top_products = df.groupby('Description')['TotalSales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
country_sales = df.groupby('Country')['TotalSales'].sum().sort_values(ascending=False)
print(country_sales.head(10))
```

```python
df.to_csv("OnlineRetail_Cleaned.csv", index=False)
```

```python
# Filter transactions with unusually high quantity (outliers)
outliers = df[df['Quantity'] > 1000]
print("Suspicious transactions (Quantity > 1000):")
print(outliers.head())

# Optionally, drop or keep separately for investigation
df = df[df['Quantity'] <= 1000]
```

```python
# Compute total revenue per transaction
df['Revenue'] = df['Quantity'] * df['UnitPrice']

# Sort descending by Revenue
df_sorted = df.sort_values(by='Revenue', ascending=False)
print(df_sorted.head())
```

```python
# Example custom data
product_info = pd.DataFrame({
    'StockCode': ['85123A', '71053', '84406B', '84029G', '22752'],
    'Category': ['Home Decor', 'Toys', 'Kitchen', 'Giftware', 'Stationery'],
    'Supplier': ['ABC Ltd', 'XYZ Imports', 'GlobalMart', 'StarSupply', 'PaperWorks']
})

# Merge with main dataset
df_merged = df.merge(product_info, on='StockCode', how='left')
print(df_merged.head())
```

```python
country_summary = df.groupby('Country').agg(
    Total_Revenue=('Revenue', 'sum'),
    Mean_Revenue=('Revenue', 'mean'),
    Order_Count=('InvoiceNo', 'nunique')
).sort_values(by='Total_Revenue', ascending=False)

print(country_summary.head(10))
```

```python
# Define bins and labels
bins = [0, 2, 10, 50, np.inf]
labels = ['Cheap', 'Medium', 'Expensive', 'Premium']

# Categorize products based on UnitPrice
df['Price_Range'] = pd.cut(df['UnitPrice'], bins=bins, labels=labels)

# Group by Price Range
price_summary = df.groupby('Price_Range').agg(
    Purchases=('InvoiceNo', 'count'),
    Avg_Revenue=('Revenue', 'mean')
)

print(price_summary)
```

```python
# Correlation among numeric columns
corr = df[['Quantity', 'UnitPrice', 'Revenue']].corr()
print(corr)
```
